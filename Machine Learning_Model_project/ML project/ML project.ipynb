```python
college_studentdatas['Prev_Sem_Result'].fillna(7.5,inplace=True)
college_studentdatas['CGPA'].fillna(7.55,inplace=True)
college_studentdatas['Academic_Performance'].fillna(5,inplace=True)
college_studentdatas['Internship_Experience'].fillna(1,inplace=True)
college_studentdatas['Extra_Curricular_Score'].fillna(4,inplace=True)
college_studentdatas['Communication_Skills'].fillna(6,inplace=True)
college_studentdatas['Projects_Completed'].fillna(3,inplace=True)
college_studentdatas['Placement'].fillna(0,inplace=True)
```

```python
college_studentdatas.isnull().sum()
```

```python
A=college_studentdatas.drop(['College_ID','IQ'],axis=1)
B=college_studentdatas['Placement']
A
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(A, B, test_size=0.2, random_state=60)
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
prediction=model.predict(X_test)
prediction
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test, prediction) 
acc
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(Y_test, prediction)
cm = cm[:3, :2] 
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')


plt.xlabel('Placed Students')
plt.ylabel('Not Placed Students')
plt.title('Confusion Matrix')
plt.show()
```
