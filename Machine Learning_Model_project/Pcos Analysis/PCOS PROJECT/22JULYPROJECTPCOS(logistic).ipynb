```python
X4 = p2.drop(["Antral_Follicle_Count","PCOS_Diagnosis"], axis=1)
Y4 = p2["PCOS_Diagnosis"]
X4,Y4
```

```python
from sklearn.model_selection import train_test_split
X4_train,X4_test,Y4_train,Y4_test=train_test_split(X4,Y4,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
p3 = LogisticRegression()
p3.fit(X4_train, Y4_train)
```

```python
X4_test
```

```python
X4_train
```

```python
Y4_test
```

```python
Y4_train
```

```python
y_pred = p3.predict(X4_test)
y_pred
```

```python
Y4_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(Y4_test, y_pred) 
print(accuracy)
accuracy*100
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(Y4_test, y_pred)
plt.figure(figsize=(4, 2))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```
