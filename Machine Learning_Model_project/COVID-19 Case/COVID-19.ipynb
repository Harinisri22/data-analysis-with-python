```python
import numpy as np
import matplotlib.pyplot as plt

x = covid_19["Name of State / UT"]
y = covid_19["New cases"]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='darkblue', alpha=0.6)

plt.xlabel("States/UT")
plt.ylabel("New COVID-19 Cases")
plt.title("COVID-19 New Case Distribution by State")

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
state_cases = covid_19.groupby("Name of State / UT")["New cases"].sum().reset_index()

plt.figure(figsize=(8, 5))
plt.plot(state_cases["Name of State / UT"], state_cases["New cases"], marker='s', color='red')

plt.xlabel("States/UT")
plt.ylabel("Total New Cases")
plt.title("COVID-19 Case Count by State")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

states = np.array(["Tamil Nadu", "Kerala", "Karnataka", "Delhi", "Maharashtra"])
cases = np.array([2300, 1800, 2100, 1500, 3000])

plt.figure(figsize=(6, 5))
plt.bar(states, cases, color='pink')

plt.title("COVID-19 Cases by State", fontsize=12)
plt.xlabel("States", fontsize=12)
plt.ylabel("Number of Cases", fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

sizes = np.array([250, 100, 75, 50])
labels = ["Recovered", "Deaths","Active Cases", "Hospitalized"]
explode = [0.1, 0.1, 0.05, 0]  

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title("COVID-19 Case Outcome")
plt.axis('equal') 
plt.show()
```
